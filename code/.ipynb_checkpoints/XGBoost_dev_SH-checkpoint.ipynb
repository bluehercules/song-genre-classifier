### 1. Training XGBoost Model

- Normalisation/standardisation not required

REFERENCE: https://www.kaggle.com/code/emmanuelfwerr/xgboost-multi-class-classification

```python
# Split train and test data
X = train.drop('genre', axis=1).copy()
y = train['genre'].copy().map(genres)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

#### 1.1 Baseline Model (FIRST XGBOOST MODEL SUBMISSION)

```python
xgb_clf = xgb.XGBClassifier(objective='multi:softmax', 
                            num_class=4, 
                            alpha=1, 
                            learning_rate=0.1,
                            max_depth=4,
                            eval_metric=['merror','mlogloss'],
                            seed=42)
xgb_clf.fit(X_train, 
            y_train,
            verbose=0,
            eval_set=[(X_train, y_train), (X_test, y_test)])

y_pred = xgb_clf.predict(X_test)
```

```python
# preparing evaluation metric plots
results = xgb_clf.evals_result()
epochs = len(results['validation_0']['mlogloss'])
x_axis = range(0, epochs)
```

```python
## CV Score
np.mean(cross_val_score(xgb_clf, X, y, cv = 5, error_score="raise"))
# 0.6408 XGB1
```

```python
## ---------- Model Classification Report ----------
## get predictions and create model quality report

print('\n--------------- Classification Report ---------------\n')
print(classification_report(y_test, y_pred))
print('---------------------- XGBoost ----------------------') # unnecessary fancy styling
```

#### 2. New Ensemble Approach

Idea is to first predict if rap/hip hop vs rock/pop. Then split the data accordingly and then predict which is which.

```python
# Set values for hyperparam optimisation
cv_params = {
    # Model complexity
    'max_depth' : np.arange(1,10,1),
    'min_child_weight' : [1,5,15,200],
    'gamma' : np.arange(0,0.5, 0.1),
    # Robustness to noise
    'subsample' : np.arange(0,1,0.1),
    'colsample_bytree' : np.arange(0,1,0.1)
            }

# Fixed parameters
fix_params_bin = {
              'learning_rate': 0.22,
              'objective': 'binary:logistic',
              'eval_metric' : ['rmse','auc'],
              'seed' : 42}

# Configure randomized search
rs_bin = RandomizedSearchCV(estimator=XGBClassifier(**fix_params_bin),
                        param_distributions=cv_params,
                        cv=5,
                        n_iter = 10)
```

```python
### Model to split rock&pop from rap&hiphop
```

```python
# Integer encoding target variable 
genres_split = {'pop': 1,
          'hip hop': 0,
          'rock': 1,
          'rap':0
         }

# Split train and test data
X_sp = rescaled_train.drop('genre', axis=1).copy()
y_sp = rescaled_train['genre'].copy()

X_train_split, X_test_split, y_train_split, y_test_split_original = train_test_split(X_sp, y_sp, random_state=42)
y_train_split = y_train_split.map(genres_split)
y_test_split = y_test_split_original.map(genres_split)
```

```python
# Split model - Hyperparam optimisation
search_bin = rs_bin.fit(X_sp, y_sp.map(genres_split))
best_params_bin = search_bin.best_params_
gen_params_bin = fix_params_bin | best_params_bin
print(gen_params_bin)
```

```python
# Train split Model
xgb_split = xgb.XGBClassifier(**gen_params_bin)
xgb_split.fit(X_train_split, 
            y_train_split,
            verbose=0,
            eval_set=[(X_train_split, y_train_split), (X_test_split, y_test_split)])
```

```python
y_pred_split = xgb_split.predict(X_test_split)

print('\n--------------- Classification Report ---------------\n')
print(classification_report(y_test_split, y_pred_split))
```

```python
np.mean(cross_val_score(xgb_split, X_sp, y_sp.map(genres_split), cv = 2, error_score="raise"))
```

#### 1.3 Assess feature importance for each genre

```python
def genxgb(gen):
    # Integer encoding target variable
    
    genres = {'pop': 0,
          'hip hop': 0,
          'rock': 0,
          'rap':0
         }
    genres[gen] = 1
    X = train.copy().drop('genre', axis=1)
    y = train.copy()['genre'].map(genres)
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    
    # Base model without hyperparameter tuning
    xgb_clf = xgb.XGBClassifier(objective='binary:logistic', 
                            missing=1, 
                            early_stopping_rounds=10, 
                            eval_metric=['error','auc'], 
                            seed=42)
    xgb_clf.fit(X_train, 
            y_train,
            verbose=0, # set to 1 to see xgb training round intermediate results
            eval_set=[(X_train, y_train), (X_test, y_test)])
    
    return xgb_clf
```

```python
gens = ['pop','hip hop','rock','rap']

for g in range(len(gens)):
    fig, ax = plt.subplots(figsize=(10,6))
    plot_importance(genxgb(gens[g]),ax=ax,max_num_features=20)
    ax.set_title(f'XGBoost Binary Logistic Feature Importance - {gens[g]}')
    plt.grid(False)
    plt.show()
    plt.gcf().subplots_adjust(left=0.2)
    fig.savefig(f'../writeup/Figures/XGBFeatImp{gens[g]}.jpg');
```
